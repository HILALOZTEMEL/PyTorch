**Max-Pooling   Oop**

```python
import torch
import torch.nn

im= torch .Tensor([[[[3,1,3,5],[6,0,7,9],[3,2,1,4],[0,2,4,3]]]])

max_pooling=torch.nn.MaxPool2d(2)
output_feature=max_pooling(im)
print(output_feature)
```

**Max-Pooling Functional**

```python
import torch
import torch.nn.functional as F

im= torch .Tensor([[[[3,1,3,5],[6,0,7,9],[3,2,1,4],[0,2,4,3]]]])
output_feature_F=F.max_pool2d(im,2)
print(output_feature_F)
```

görüntü 4 boyuta sahiptir.(minibatch boyutu, derinlik,yükseklik ve genişlik için) daha sonra çekirdek boyutu 2 olan maksimum havuzlama nesnelerini oluşturuyoruz ve torç kullandığımızda bu görüntümüze uyguluyoruz.

**Average pooling in PyTorch**

```python
#oop
import torch
import torch.nn

im= torch .Tensor([[[[3,1,3,5],[6,0,7,9],[3,2,1,4],[0,2,4,3]]]])

avg_pooling=torch.nn.AvgPool2d(2)
output_feature=avg_pooling(im)
print(output_feature)
```

```python
#functional
import torch
import torch.nn.functional as F

im= torch .Tensor([[[[3,1,3,5],[6,0,7,9],[3,2,1,4],[0,2,4,3]]]])
output_feature_F=F.avg_pool2d(im,2)
print(output_feature_F)
```
